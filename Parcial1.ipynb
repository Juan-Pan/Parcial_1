```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns

# Recoleccion de datos
from google.colab import drive
drive.mount ("/content/drive")

Casos_covid = pd.read_csv("/content/drive/MyDrive/Casos_positivos_de_COVID-19_en_Colombia._20250311.csv")

# Eliminar duplicados
Casos_covid.drop_duplicates(inplace = True)

# Convertir edades negativas a positivas
Casos_covid["Edad"] = Casos_covid["Edad"].apply(lambda x : abs(x) if x < 0 else x)

# Corregir valores incorrectos en la columna "Sexo"
Casos_covid["Sexo"] = Casos_covid["Sexo"].str.strip().str.upper()

# Convertir a mayusculas y eliminar espacios en blanco
columnas_fecha = ["fecha reporte web", "Fecha de notificación", "Fecha de inicio de síntomas", "Fecha de diagnóstico", "Fecha de recuperación"]

for columna in columnas_fecha :
  if columna in Casos_covid.columns :
    Casos_covid[columna] = pd.to_datetime (Casos_covid[ columna ], errors="coerce")
  else :
    print(f"La columna '{columna}' no existe en el DataFrame.")

# Procesamiento de los datos
data_frame=pd.DataFrame(Casos_covid)

# Guardar los datos limpios en un nuevo archivo
Casos_covid.to_csv("datos_limpiosarch2.csv", index=False )
# Se guarda los datos en un nuevo archivo llamado "datos_limpios2.csv"
```

```python
def understanding_data(data_frame):
    print("Number of rows::", data_frame.shape[0])
    print("\nNumber of columns::", data_frame.shape[1])
    print("\nColumn names::", data_frame.columns.values.tolist())

    # Identificar y mostrar las columnas que contienen valores nulos
    print("\nColumns with Missing Values::", data_frame.columns[data_frame.isnull().any()].tolist())

    # Contar y mostrar el número de filas que contienen al menos un valor nulo
    print("\nNumber of rows with Missing Values::", len(data_frame[data_frame.isnull().any(axis=1)]))

    # Mostrar los índices de algunas filas que contienen valores nulos (máximo 20)
    print("\nSample Indices with missing data::", data_frame[data_frame.isnull().any(axis=1)].index.tolist()[0:20])

    # Estadistica generales del dataframe
    print("\nGeneral Stats::")
    print(data_frame.info())
```

```python
# Columnas a eliminar del datafreame
Columns_TD = ['Código ISO del país', 'Nombre departamento', 'Nombre del grupo étnico', 'Nombre del país','Pertenencia étnica','Fecha de muerte', 'ID de caso', 'Código DIVIPOLA departamento','Código DIVIPOLA municipio']

# Función principal para imputar valores faltantes en el dataframe
def impute_missing_values(data_frame):
    data_frame = delete_columns(data_frame)
    data_frame = typecast(data_frame)
    data_frame = fill_age(data_frame)
    data_frame = fill_gender(data_frame)

    # Llenar valores nulos en otras columnas
    empty_values(data_frame)
    data_frame = uppercase(data_frame)
    return data_frame

# Función para eliminar columnas innecesarias
def delete_columns(data_frame):
    for column in Columns_TD:
        if column in data_frame.columns:
            data_frame.drop(column, axis=1, inplace=True)
            print(f"Column '{column}' deleted successfully.")
        else:
            print(f"Column '{column}' not found in the DataFrame.")
    return data_frame

# Función para convertir tipos de datos en columnas específicas (typecasting)
def typecast(data_frame):
    data_frame['Fecha de notificación'] = pd.to_datetime(data_frame['Fecha de notificación'])
    data_frame['Fecha de inicio de síntomas'] = pd.to_datetime(data_frame['Fecha de inicio de síntomas'])
    data_frame['Fecha de diagnóstico'] = pd.to_datetime(data_frame['Fecha de diagnóstico'])
    data_frame['Fecha de recuperación'] = pd.to_datetime(data_frame['Fecha de recuperación'])
    data_frame['Edad'] = data_frame['Edad'].astype(int)

    return data_frame

# Función para llenar valores faltantes en la columna "Edad" con la media
def fill_age(data_frame):
    data_frame['Edad'] = data_frame['Edad'].fillna(data_frame['Edad'].mean())
    return data_frame

# Función para llenar valores faltantes en la columna "Sexo" con la moda
def fill_gender(data_frame):
    data_frame['Sexo'] = data_frame['Sexo'].fillna(data_frame['Sexo'].mode()[0])
    return data_frame

# Función para llenar valores nulos en todas las columnas
def empty_values(data_frame):
    for column in data_frame.columns:
        if data_frame[column].dtype == 'object':
            data_frame[column] = data_frame[column].fillna(data_frame[column].mode()[0])
        else:
            data_frame[column] = data_frame[column].fillna(data_frame[column].mean())

    return data_frame

# Función para convertir los valores de texto a mayúsculas
def uppercase(data_frame):
    for column in data_frame.columns:
        if data_frame[column].dtype == 'object':
            data_frame[column] = data_frame[column].str.upper()

    return data_frame
```

```python
sns.kdeplot(data_frame['Edad'], fill=True, color='green')

plt.xlabel('Edad')
plt.ylabel('Densidad')
plt.title('Densidad de edades en pacientes con COVID-19')

mediana_edad = data_frame['Edad'].median()
media_age = data_frame['Edad'].mean()

plt.axvline(x=media_age, color='b', linestyle='--', label=f'Media: {media_age:.1f} años')
plt.axvline(x=mediana_edad, color='r', linestyle='--', label=f'Mediana: {mediana_edad:.1f} años')

plt.legend()
# Visualizacion del grafico
plt.grid(True)
plt.show()
```
